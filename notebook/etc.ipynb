```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage
from skimage import io
```

```python
data = io.imread('/Users/steenbender/Desktop/PhD/Funddamentals PhD/Module 3/Convolution_lesson/data/img/frog.jpg')

fig, ax = plt.subplots(1 , figsize=(10, 10))
ax.imshow(data)
red_box = plt.Rectangle((225, 375), 25, 25, edgecolor='red', facecolor='none', linewidth=2)
ax.add_patch(red_box)
ax.axis('off')
```

```python
# zoom into a section of the image
x_start, x_end = 225, 250
y_start, y_end = 375, 400
zoomed_data = data[y_start:y_end, x_start:x_end]
```

```python
fig, ax = plt.subplots(1 , figsize=(5, 5))
ax.imshow(zoomed_data)
ax.axis('off')
```

```python
fig, ax = plt.subplots(1,3,figsize=(25, 30))
# Show all the three channels with the value of the pixel printed in the center
ax[0].imshow(zoomed_data[:,:,0], cmap='Reds')
for i in range(zoomed_data.shape[0]):
    for j in range(zoomed_data.shape[1]):
        ax[0].text(j, i, zoomed_data[i,j,0], ha='center', va='center', color='black')
ax[0].set_title('Red Channel')
ax[1].imshow(zoomed_data[:,:,1], cmap='Greens')
for i in range(zoomed_data.shape[0]):
    for j in range(zoomed_data.shape[1]):
        ax[1].text(j, i, zoomed_data[i,j,1], ha='center', va='center', color='black')
ax[1].set_title('Green Channel')
ax[2].imshow(zoomed_data[:,:,2], cmap='Blues')
for i in range(zoomed_data.shape[0]):
    for j in range(zoomed_data.shape[1]):
        ax[2].text(j, i, zoomed_data[i,j,2], ha='center', va='center', color='black')
ax[2].set_title('Blue Channel')
[x.axis('off') for x in ax]
plt.tight_layout()
```
